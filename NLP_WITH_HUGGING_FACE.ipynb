```python
from transformers import pipeline

qa_pipeline = pipeline("question-answering", model="distilbert-base-cased-distilled-squad")
context = "Singa adalah hewan karnivora yang dikenal sebagai raja hutan."
question = "Apa itu singa?"

result = qa_pipeline(question=question, context=context)
print(f"Jawaban: {result['answer']}")
```

```python
from transformers import pipeline

qa_pipeline = pipeline("question-answering", model="distilbert-base-cased-distilled-squad")
context = "Albert Einstein, lahir pada 14 Maret 1879, adalah seorang fisikawan teoritis yang dikenal karena teori relativitasnya, yang mengubah cara kita memahami ruang dan waktu."
question = "Tanggal berapa Albert Einstein lahir?"

result = qa_pipeline(question=question, context=context)
print(f"Jawaban: {result['answer']}")
```

```python
from transformers import pipeline

summarizer = pipeline("summarization")
summarizer(
    """
    The digital revolution has fundamentally changed the way we communicate, work,
    and live. With the rise of the internet, information has become more accessible than
    ever before. Social media platforms have transformed personal interactions, allowing
    people to connect across the globe instantly. However, this revolution has also
    brought challenges, such as the spread of misinformation and concerns about privacy.
    In the workplace, automation and artificial intelligence are reshaping job roles,
    enhancing productivity but also raising fears of job displacement.

    Education systems are also adapting to this new landscape, integrating technology
    into curricula to prepare students for a digital future. As we navigate these changes,
    it is crucial to find a balance between embracing innovation and addressing the ethical
    implications that come with it. Policymakers, educators, and businesses must work together
    to ensure that the benefits of the digital age are shared equitably while safeguarding
    individual rights and societal values.
"""
)
```

```python
from transformers import pipeline

translator = pipeline("translation", model="Helsinki-NLP/opus-mt-id-en")

text_to_translate = "Saya senang berlari di pagi hari."
result = translator(text_to_translate)

print(result[0]['translation_text'])
```
